```python
from scipy.constants import physical_constants
import numpy as np
import matplotlib.pyplot as plt
import sys
```

```python
T0=10
T1=300
emin=-2.0
emax=2.0
mu0=0.5
mu1=0.5
nint=1000
#### CONSTANTS
kb=physical_constants['Boltzmann constant in eV/K'][0]
#hartree=physical_constants['hartree-electron volt relationship'][0]
#kb=kb/hartree
#### CONSTANTS
fig,ax1=plt.subplots(1,1,figsize=(10,6))

lab0='T= {:6.2f} K'.format(T0)
lab1='T= {:6.2f} K'.format(T1)
#ax1.set_aspect(1)
ax1.axes.set_xlim([-2.0,2])
ax1.axes.set_ylim([-0.1,1.1])
ax1.set_title('Fermi-Dirac distribution $\mu = 0.5 eV$',size=20)
ax1.tick_params(labelsize=20)
ax1.set_xlabel(r'$E \, \, in \, \,  eV$',fontsize=20)
ax1.set_ylabel('occupation number',fontsize=20)
xp0=[]
yp0=[]
yp1=[]
for i in range(1000):
    x=emin + (emax-emin)/(nint-1)*i
    xp0.append(x)
    y=1.0/(np.exp((x-mu0)/(kb*T0))+1)
    y1=1.0/(np.exp((x-mu1)/(kb*T1))+1)
    yp0.append(y)
    yp1.append(y1)

ax1.plot(xp0,yp0,label=lab0,color='b',linewidth=3,linestyle='-')
ax1.plot(xp0,yp1,label=lab1,color='r',linewidth=3,linestyle='-')
ax1.legend(loc="upper right")
fig.savefig("fermi.png", dpi=300)
plt.show()
```
